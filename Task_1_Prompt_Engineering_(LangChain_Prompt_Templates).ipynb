**Task 4: ChatPromptTemplate System**

```python
# Task 4: ChatPromptTemplate

def get_chat_prompt(role, topic):

    system_roles = {
        "teacher": "You are a helpful teacher who explains concepts clearly.",
        "interviewer": "You are an interviewer asking challenging questions.",
        "motivator": "You are a motivational speaker who inspires learning."
    }

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_roles.get(role, "You are a helpful assistant.")),
        ("user", "Tell me about {topic}")
    ])

    return chat_template.format_messages(topic=topic)


# Example
messages = get_chat_prompt("teacher", "Neural Networks")
for msg in messages:
    print(msg)
```

**Task 5: Input Validation Layer**

```python
# Task 5: Validation function

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Defaulting to 'formal'")
        tone = "formal"

    return audience, tone


# Test
aud, tone = validate_inputs("kids", "funny")
print(aud, tone)
```

**Task 6: Prompt Generator App**

```python
# Task 6: Prompt Generator Function

def generate_prompt(topic, audience, tone, style):

    # Validate inputs
    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates.get(style, templates["teaching"])
    )

    return prompt_template.format(topic=topic, audience=audience, tone=tone)


# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**Task 7: Template Reusability Test**

```python
# Task 7: Reusability with one template

reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

# 5 different inputs
inputs = [
    ("AI", "beginners"),
    ("Python", "students"),
    ("Data Science", "analysts"),
    ("Deep Learning", "engineers"),
    ("Cloud Computing", "professionals")
]

for topic, audience in inputs:
    print(reusable_template.format(topic=topic, audience=audience))
```
